```python
# importing
import pandas as pd
import missingno as miss
```

```python
# read in the data
data = pd.read_csv('allCongressDataPublishV2.csv')
```

```python
# get the full list of columns 
data.columns
```

```python
# create a list of variables of interest
variables = ['state', 'district','totalHouseholds', 'medianIncome', 'prcntBA', 'prcntHS',
       'prcntAsian', 'prcntBlack', 'prcntBlackNotHisp', 'prcntHisp',
       'prcntMulti', 'prcntWhite', 'prcntWhiteAll', 'prcntNotHisp', 'prcntOld',
       'medianAge', 'lastName','firstName', 'middleName', 'age', 'gender', 'numberTerms',
       'daysServed', 'fracServed', 'cd', 'statenm', 'dwnom1', 'dwnom2',
       'comPower', 'chair', 'rank', 'comName', 'numCom', 'rankChair', 'black',
       'hispanic', 'numSpon', 'numCosp', 'numPassH', 'numEnact', 'passPrcnt',
       'congNum', 'partyControl']
```

```python
# subset the data frame
data = data[variables]
```

```python
# show matrix of missingness with missing no
miss.matrix(data)
```

```python
# we remove the variables with lots of missingness
data.drop(columns = ['medianIncome', 'prcntBlackNotHisp','prcntWhite','prcntNotHisp','prcntOld','prcntMulti','medianAge','comPower','chair','rank','comName','numCom','rankChair'],inplace=True)
```

```python
# show matrix of missingness with missing no
miss.matrix(data)
```

```python
# output the data
data.to_csv('data/pre_made_wrangled.csv', encoding='utf-8', index=False)
```
