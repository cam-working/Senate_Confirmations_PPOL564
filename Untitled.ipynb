```python
set(skel['District'].items())
```

```python
district_string_to_number = {
    'At-large' :'0.0',
    'At large' :'0.0',
}
```

```python
skel = skel.replace({'District': district_string_to_number})
```

```python
skel = skel.replace({'State': us_state_to_abbrev})
```

```python
skel['District'] = skel['District'].str.replace('th','.0')
skel['District'] = skel['District'].str.replace('rd','.0')
skel['District'] = skel['District'].str.replace('st','.0')
skel['District'] = skel['District'].str.replace('nd','.0')
```

```python
skel['District'] = skel['District'].astype('float')
```

```python
skel.set_index(['State','District','Congress'], inplace = True)
```

```python
addition.rename(columns={'congNum':'Congress', 'state':'State', 'district':'District'}, inplace=True)
```

```python
addition.set_index(['State','District','Congress'], inplace=True)
```

```python
final = skel.join(addition)
```

```python
miss.matrix(final)
```

```python
data = final.dropna().reset_index()
```

```python
data.to_csv('data.csv', encoding='utf-8', index=False)
```
